```python
#共通前処理

#余分なワーニングを消す
import warnings
warnings.filterwarnings('ignore')

#ライブラリのインポート
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib

#データフレーム表示用関数
from IPython.display import display

#npでの浮動小数点の表示制精度
np.set_printoptions(suppress=True, precision=4)

#pdでの浮動小数点の表示精度
pd.options.display.float_format = '{:.4f}'.format

#データフレームですべての項目を表示
pd.set_option('display.max_columns', None)

#グラフのフォント指定
plt.rcParams['font.size'] = 14

#乱数
random_seed = 123
```

```python
#追加ライブラリのimprt

import seaborn as sns

#データの読み込み
df_titanic = sns.load_dataset('titanic')

#項目の日本語化
columns_t = ['生存', '等室', '性別', '年齢', '兄弟配偶者数', 
             '両親子供数', '料金', '乗船港コード', '等室名', 
             '男女子供', '成人男子', 'デッキ', '乗船港', '生存可否', '独身']

df_titanic.columns = columns_t

#でーたないようの確認
display(df_titanic.head())
```

```python
#欠損値がある項目を調べる
print(df_titanic.isnull().sum())
```

```python
#乗船港の項目値ごとの個数
print(df_titanic['乗船港'].value_counts())
print()

#生存可否の項目値ごとの個数
print(df_titanic['生存可否'].value_counts())
```

```python
#統計情報を調べるdescribe関数で行ける
display(df_titanic.describe())
```

```python
#性別別に平均値を確認する
display(df_titanic.groupby('性別').mean())
```

```python
#分析対象の項目をグラフ表示(数値項目の場合)

#項目の定義
columns_n = ['生存', '等室', '年齢', '兄弟配偶者数', '両親子供数', '料金' ]

#グラフ領域の調整
plt.rcParams['figure.figsize'] = (10, 10)

#ヒストグラムを表示する
df_titanic[columns_n].hist()
plt.show()
```

```python
#分析対象の項目をグラフ表示(非数値項目の場合)

#グラフ化対象列の定義

columns_c = ['性別', '乗船港', '等室名','成人男子']

#グラフ領域の調整
plt.rcParams['figure.figsize'] = (8, 8)

#ループ処理でヒストグラムの表示
for i,name in enumerate(columns_c):
    ax = plt.subplot(2,2, i+1)
    df_titanic[name].value_counts().plot(kind='bar', title=name, ax=ax)

#レイアウトの調整
plt.tight_layout()
plt.show()
```

```python
#追加ライブラリのインポート
import seaborn as sns

#データの読み込み
df_iris = sns.load_dataset('iris')

#項目の日本語化
columns_i = ['がく片長', 'がく片幅', '花弁長', '花弁幅', '種別']

#カラム名の変更
df_iris.columns = columns_i

#データ内容の出力
display(df_iris.head())
```

```python
#matplotで散布図を表示する

plt.figure(figsize=(6,6))

#散布図の表示
plt.scatter(df_iris['がく片幅'], df_iris['花弁長'])

#ラベルの表示
plt.xlabel('がく片幅')
plt.ylabel('花弁長')
plt.show()
```

```python
#seabornを使った散布図の表示

#hueは色分け

plt.figure(figsize=(6,6))
sns.scatterplot(x='がく片幅', y='花弁長', hue='種別', s=70, data=df_iris)
plt.show()
```

```python
#seabornのpairplot関数による散布図の表示
#きれいにグループ分けできているかを確認することができる

sns.pairplot(df_iris, hue='種別')
plt.show()
```

```python
#joinplot関数による散布図の表示
#散布図の外側にヒストグラムが表示される

sns.jointplot('がく片幅', '花弁長', data= df_iris)
plt.show()
```

```python
#matplotを使用した箱ひげ図お表示
#patch_artistは箱を塗りつぶすかどうかを指定している

#グラフ領域の調整
plt.figure(figsize=(6,6))

#箱ひげ図の作成
df_iris.boxplot(patch_artist=True)
plt.show()
```

```python
#serbornで箱ひげ図を表示する

#melt関数でデータの事前加工
w = pd.melt(df_iris, id_vars=['種別'])

#加工結果の確認
display(w.tail())
```

```python
#serbornで箱ひげ図を表示する

#hueパラメータを使い、花の種類ごとに箱ひげ図を書いていく
#対象データは上で加工処理を加えたデータ

plt.figure(figsize=(8,8))
sns.boxplot(x='variable', y='value',data=w, hue='種別')
plt.show()
```
