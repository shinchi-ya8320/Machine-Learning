```python
#警告を非表示にするためのやつ
import warnings
warnings.filterwarnings('ignore')

#必須なライブラリ
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#matplotlibを日本語化
import japanize_matplotlib

#データフレーム表示用の関数
from IPython.display import display

#表示オプションの調整

#浮動小数点の表示制度
np.set_printoptions(suppress=True, precision=4)

#浮動小数点の表示制度
pd.options.display.float_format = '{:.4f}'.format

#グラフのデフォルトフォント設定
plt.rcParams['font.size'] = 14

#乱数の種類
random_seed = 123
```

```python
#データセットのロード

#ライブラリのインポート
from sklearn.datasets import load_breast_cancer

#データのロード
cancer = load_breast_cancer()

#データの注釈を読む
print(cancer.DESCR)
```

```python
#データフレームへの取り込み

columns = [
    '半径_平均', 'きめ_平均', '周長_平均', '面積_平均', 
    '平滑度_平均','コンパクト度_平均', '凹面_平均',
    '凹点_平均', '対称性_平均', 'フラクタル度_平均',
    '半径_標準誤差', 'きめ_標準誤差', '周長_標準誤差',
    '面積_標準誤差', '平滑度_標準誤差',
    'コンパクト度_標準誤差', '凹面_標準誤差', '凹点_標準誤差',
    '対称性_標準誤差', 'フラクタル度_標準誤差',
    '半径_最大', 'きめ_最大', '周長_最大', '面積_最大', 
    '平滑度_最大','コンパクト度_最大', '凹面_最大', '凹点_最大', 
    '対称性_最大', 'フラクタル度_最大'
]

#ロードしたデータのデータフレームの取り込み
df = pd.DataFrame(cancer.data, columns=columns)

#正解データの取得
y = pd.Series(cancer.target)
```

```python
#上まででデータの読み込みが完了したのでデータの確認を行う

#入力データの表示(入力データの先頭から20~24行目までを表示する)print関数でも行けるがこっちのほうがきれいにでる
display(df[20:25])
```

```python
#正解データの表示
print(y[20:25])
```

```python
#データの統計情報を確認する

#入力データの行数と列数を取得
print(df.shape)

#正解データの1と0の個数を確認する
print(y.value_counts())
```

```python
#散布図を描く
#正解データを0のグループと1のグループに分ける

#正解データ = 0 (悪性データ)の抽出
df0 = df[y==0]

#正解データ = 0 (良性データ)の抽出
df1 = df[y==1]

display(df0.head())
display(df1.head())
```

```python
#散布図の表示

#グラフのサイズ設定
plt.figure(figsize=(6,6))

#目的変数が0のデータを散布図に
plt.scatter(df0['半径_平均'], df0['きめ_平均'], marker='x', c ='b', label = '悪性')

#目的変数が1のデータを散布図に
plt.scatter(df1['半径_平均'], df1['きめ_平均'], marker='s', c ='k', label = '良性')

#格子
plt.grid()

#ラベル
plt.ylabel('半径_平均')
plt.xlabel('きめ_平均')

#凡例
plt.legend()

#グラフ表z
plt.show()

```

```python
input_columns = ['半径_平均', 'きめ_平均']
x = df[input_columns]
display(x.head())
```

以下は訓練データと検証データの分割を行う
訓練データと検証データを7:3の割合で分割する

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test, = train_test_split(x, y, train_size=0.7, test_size=0.3, random_state=random_seed)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
display(x_train.head())
display(x_test.head())
display(y_train.head())
display(y_test.head())
```

```python
#アルゴリズムの選択
from sklearn.linear_model import LogisticRegression
algorithm = LogisticRegression(random_state=random_seed)
```

```python
#学習
algorithm.fit(x_train, y_train)
print(algorithm)
```

```python
#予測

#predict関数の呼び出し
y_pred = algorithm.predict(x_test)
print(y_pred)
```

```python
#正解データと予測結果の比較

#正解データ　先頭から10こ
#y_testはデータフレームなのでvaluesでnumpyへ変換する
y_test10 = y_test[:10].values
print(y_test10)

#予測結果　先頭から0こ
y_pred10 = y_pred[:10]
print(y_pred10)
```

```python
#正解数のカウント

#正解データと予測結果が一致するものをtrue,しないものをfalseで返す。numpy配列が返される
w1 = (y_test10 == y_pred10)
print(w1)

#正解データの数を合計する。trueが1、falseが0
w2 = w1.sum()
print(w2)
```

```python
#正解数のカウント

#正解データと予測結果が一致するものをtrue,しないものをfalseで返す。numpy配列が返される
#wには検証データすべてが格納されている
w = (y_test == y_pred)
#正解データの計算
correct = w.sum()

N = len(w)

#精度 = 正解数/検証データの全体数
score = correct / N

print(f'制度： {score:.04f}')
```

```python
#score関数の利用モデルの精度を計算できる関数
score = algorithm.score(x_test, y_test)
print(f'score: {score:.04f}')
```

```python
#モデルの制度をチューニングして上げる

#オリジナルの30項目の入力データを使用して訓練データと検証データを作り出す
x2_train, x2_test, y_train, y_test = train_test_split(df, y, train_size=0.7, test_size=0.3, random_state=random_seed)

#モデルのインスタンスを作り直す
algorithm2 = LogisticRegression(random_state=random_seed)

#訓練データで学習する
algorithm2.fit(x2_train, y_train)

#精度の確認
score2 = algorithm2.score(x2_test, y_test)

print(f'score: {score2:.04f}')
```
