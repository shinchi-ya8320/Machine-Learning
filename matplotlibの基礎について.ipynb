#### matplotlibの操作の基礎について  

パーツに当てはめてグラフを描画する  

1,figure(台紙)を用意する  
2,axes(どんなグラフを書くのか)を指定する  
3,その他オプション

```python
from matplotlib import pyplot as plt
import numpy as np
import japanize_matplotlib
```

```python
#figsizeは台紙のサイズを指定する場所。指定しなくても行ける
fig = plt.figure(figsize=(5,5))
ax = plt.axes()
plt.show()

#台紙とその枠を表示しただけ
```

```python
print(np.random.rand(1,10, (1)))
```

```python
#グラフを描画するためのダミーデータを作成する

x = [105, 210, 301, 440, 500]

y = [10, 20, 30, 40, 50]
```

### 折れ線グラフ

```python
#figsizeは台紙のサイズを指定する場所。指定しなくても行ける
fig = plt.figure(figsize=(5,5))
ax = plt.axes()

#折れ線グラフ
ax.plot(x, y)

plt.show()
```

```python
#figsizeは台紙のサイズを指定する場所。指定しなくても行ける
fig = plt.figure(figsize=(5,5))
ax = plt.axes()

#折れ線グラフ
ax.plot(x, y)

#グラフのタイトル
ax.set_title('検証うぇい')

#x軸とy軸のラベルを記入する
ax.set_xlabel('x')
ax.set_ylabel('y')

#グリッド線の表示
ax.grid()

#凡例の表示
#ax.legend()

plt.show()
```

```python
#折れ線の色を変えたりとかやる


#figsizeは台紙のサイズを指定する場所。指定しなくても行ける
fig = plt.figure(figsize=(5,5))
ax = plt.axes()

#折れ線グラフ
#マーカーの指定（点のこと）
#折れ線の色
#折れ線の線の種類指定
#点線にして、太さを2に変更した
ax.plot(x, y, marker='o', color = 'red', linestyle='--', linewidth=2.0)


#グラフのタイトル
ax.set_title('検証うぇい')

#x軸とy軸のラベルを記入する
ax.set_xlabel('x')
ax.set_ylabel('y')

#グリッド線の表示
ax.grid()

#凡例の表示
#ax.legend()

plt.show()
```

### 棒グラフ

```python
#棒グラフを描画するためのダミーデータを作成する

x = ['A', 'B', 'C', 'D']
y = [400, 300, 150, 300]
```

```python
#棒グラフを書く

#figsizeは台紙のサイズを指定する場所。指定しなくても行ける
fig = plt.figure(figsize=(5,5))
ax = plt.axes()

#折れ線グラフ
#マーカーの指定（点のこと）
#折れ線の色
#折れ線の線の種類指定
#点線にして、太さを2に変更した
#ax.bar(x, y, marker='o', color = 'red', linestyle='--', linewidth=2.0)
ax.bar(x, y)

#グラフのタイトル
ax.set_title('検証うぇい')

#x軸とy軸のラベルを記入する
ax.set_xlabel('x')
ax.set_ylabel('y')

#グリッド線の表示
ax.grid()

#凡例の表示
#ax.legend()

plt.show()
```

### ヒストグラム

```python
#正規分布に従う乱数を発生させる
print(np.random.randn(10))
```

```python
x = np.random.randn(1000)
```

```python
#ヒストグラムを書く

#figsizeは台紙のサイズを指定する場所。指定しなくても行ける
fig = plt.figure(figsize=(5,5))
ax = plt.axes()

#折れ線グラフ
#マーカーの指定（点のこと）
#折れ線の色
#折れ線の線の種類指定
#点線にして、太さを2に変更した
#ax.bar(x, y, marker='o', color = 'red', linestyle='--', linewidth=2.0)
ax.hist(x, bins=50)

#グラフのタイトル
ax.set_title('検証うぇい')

#x軸とy軸のラベルを記入する
ax.set_xlabel('x')

#グリッド線の表示
ax.grid()

#凡例の表示
#ax.legend()

plt.show()


#ヒストグラムの区間の幅、棒の幅の指定はbinsというパラメータですることができる
```

### 散布図の描画

```python
#ダミーデータを作成する

#1~101までのデータをランダムに200個発生させる
x = np.random.randint(1, 101, 200)
y = np.random.randint(1, 101, 200)
```

```python
#散布図を書く

#figsizeは台紙のサイズを指定する場所。指定しなくても行ける
fig = plt.figure(figsize=(5,5))
ax = plt.axes()

#折れ線グラフ
#マーカーの指定（点のこと）
#折れ線の色
#折れ線の線の種類指定
#点線にして、太さを2に変更した
#ax.bar(x, y, marker='o', color = 'red', linestyle='--', linewidth=2.0)
ax.scatter(x, y)

#グラフのタイトル
ax.set_title('検証うぇい')

#x軸とy軸のラベルを記入する
ax.set_xlabel('x')
ax.set_ylabel('y')
#グリッド線の表示
ax.grid()

#凡例の表示
#ax.legend()

plt.show()


#ヒストグラムの区間の幅、棒の幅の指定はbinsというパラメータですることができる
```

### 任意の数のグラフを1度に表示する

```python
fig = plt.figure()

#2行、2列、1番目に書く
ax1 = fig.add_subplot(221)
#2行、2列、2番目に書く
ax2 = fig.add_subplot(222)
#2行、2列、3番目に書く
ax3 = fig.add_subplot(223)
#2行、2列、4番目に書く
ax4 = fig.add_subplot(224)

plt.show()
```

```python
fig = plt.figure()

#2行、2列、1番目に書く
ax1 = fig.add_subplot(221)
#2行、2列、2番目に書く
ax2 = fig.add_subplot(222)
#2行、2列、3番目に書く
ax3 = fig.add_subplot(223)
#2行、2列、4番目に書く
ax4 = fig.add_subplot(224)

#グラフごとの縦の間隔と横の間隔を調節する
plt.subplots_adjust(wspace=0.5, hspace=0.5)

plt.show()
```

```python
#折れ線、棒グラフ、ヒストグラム、散布図を1度に表示する

fig = plt.figure(figsize=(7,7))

#2行、2列、1番目に書く
#折れ線グラフ
ax1 = fig.add_subplot(221)
ax1.set_title('ax1')
x1 = [105, 210, 301, 440, 500]
y1 = [10, 20, 30, 40, 50]
ax1.plot(x1, y1)


#2行、2列、2番目に書く
#棒グラフ
ax2 = fig.add_subplot(222)
ax2.set_title('ax2')
x2 = ['A', 'B', 'C', 'D']
y2 = [400, 300, 150, 300]
ax2.bar(x2, y2)


#2行、2列、3番目に書く
#ヒストグラム
ax3 = fig.add_subplot(223)
ax3.set_title('ax3')
x3 = np.random.randn(1000)
ax3.hist(x3, bins=50)


#2行、2列、4番目に書く
ax4 = fig.add_subplot(224)
ax4.set_title('ax4')
x4 = np.random.randint(1, 101, 200)
y4 = np.random.randint(1, 101, 200)
ax4.scatter(x4, y4)

#グラフごとの縦の間隔と横の間隔を調節する
plt.subplots_adjust(wspace=0.5, hspace=0.5)

plt.show()
```
