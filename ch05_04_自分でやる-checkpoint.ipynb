```python
#共通前処理

import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

import japanize_matplotlib

#データフレーム表示用
from IPython.display import display


#いろいろ調整
np.set_printoptions(suppress=True, precision=4)
pd.options.display.float_format = '{:.4f}'.format

#データフレームのすべての項目を表示する
pd.set_option('display.max_columns',None)

#グラフのフォント
plt.rcParams['font.size'] = 14

#乱数
random_seed = 123
```

```python
#データの読み込み
df = pd.read_excel(r'C:\Users\shoni\Downloads\Online Retail.xlsx')

#項目名の日本語化
columns = [
    '発注番号', '商品番号', '商品説明', '商品個数', '明細発行日', '商品単価','顧客番号', '国名'
]

df.columns = columns
```

```python
#欠損値の確認
print(df.isnull().sum())

#商品説明と顧客番号で欠損値があるが
#今回は特に補完の対応は必要ない
#顧客ベースでの分析はしないし、商品説明は商品辞書を作成して対応するため
```

```python
#顧客の国名の比率
print(df['国名'].value_counts().head(20))
```

データの確認が終わったのでデータの前処理  
対象を新規オーダーに限定する  
地域をフランスに限定する  
データをone hotエンコーディングへ変換する

```python
#発注情報には新規オーダーとキャンセルオーダーが含まれている
#キャンセルオーダーと分析に含めてしまうと悪い影響が出そうなので新規オーダーだけに絞る

df2 = df.copy()

#発注番号の頭文字を別項目へ
#5:新規、C:キャンセル
df2['発注種別'] = df2['発注番号'].map(lambda x: str(x)[0])

#結果
display(df2.head())

#種別個数確認
print(df2['発注種別'].value_counts())
```

```python
#新規オーだーだけを抽出する

df2 = df2[df2['発注種別']=='5']

print(df2.shape)
```

```python
#更に国をフランスへ絞る
df3 = df2[df2['国名']=='France']

print(df3.shape)
```

データをone hot エンコーディングへ直す

```python
#発注番号、商品番号の形のデータを縦持ちから横持ちへ

#発注番号と商品番号をキーに商品個数を集計する
w1 = df3.groupby(['発注番号', '商品番号'])['商品個数'].sum()

#結果
print(w1.head())
```

```python
#商品番号を列に移動する
#unstack関数で行列の移動
#fillnaで値の入っていない場所に0を入れる
w2 = w1.unstack().reset_index().fillna(0).set_index('発注番号')

#行列の確認
print(w2.shape)

#結果
display(w2.head())
```

```python
#集計結果が1以上ならtrue,0ならfalseに変換する
#apply関数はデータフレームの行列方向へlambdaを適用する
basket_df = w2.apply(lambda x: x>0)

#結果
display(basket_df.head())
```

```python
#商品名の辞書を作成する
#商品番号と商品説明だけを抽出する
w3 = df2[['商品番号', '商品説明']].drop_duplicates()

#商品番号と商品説明を文字列へ変換する
w3['商品番号'] = w3['商品番号'].astype('str')
w3['商品説明'] = w3['商品説明'].astype('str')

#商品番号をインデックスに
w3 = w3.set_index('商品番号')
display(w3.head())
```

```python
#冒頭の5つの商品コードを抽出する
item_list1 = w3.index[:5]

print(item_list1)

#w3に対する検証結果
display(w3.loc[item_list1])

#また不完全
```

```python
#大文字だけのテキストの抽出

#作業用のコピー
w4 = w3.copy()

#すべての大文字名称の行のみを残す
w4 = w4[w4['商品説明'].map(lambda x: x.isupper())]

#結果
display(w4.loc[item_list1])

#まだ重複している
```

```python
#一番長い名称の抽出

#作業用
w5 = w4.copy()

#文字数列の追加
w5['文字数'] = w5['商品説明'].map(len)

#ソート
w5 = w5.sort_values(['商品番号', '文字数'], ascending=[True, False])

#商品番号ごとに一番長い名称を抽出してitem_dictへ代入する
item_dict = w5.groupby('商品番号')['商品説明'].agg(lambda x: x[0])

#結果
display(item_dict.loc[item_list1])
```

商品の辞書を作成することができたのでアルゴリズムの選択と分析を実施する

```python
#アソシエーション分析のためのライブラリをインポートする
#アプリオリ関数とルール抽出用の関数
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#アプリオリにょる分析
#min_support = 0.06は支持度の閾値
freq_items1 = apriori(basket_df, min_support = 0.06, use_colnames=True)

#結果表示
display(freq_items1.sort_values('support', ascending=False).head(30))

#itemsetすう確認
print(freq_items1.shape)

#supportは支持度のことで0.06より大きい数字のものだけを出している
```

```python
#アソシエーションのルール抽出
#min_threshold=1はリフト値が1以上のもの

a_rules1 = association_rules(freq_items1, metric= 'lift', min_threshold=1)

#リフト値でソートする
a_rules1 =a_rules1.sort_values('lift', ascending=False).reset_index(drop=True)

#結果
display(a_rules1.head(10))

#ルール数確認
print(a_rules1.shape)
```

```python
#関係性の高い商品の商品コードをリスト化
item_list = ['23254', '23256', '22727', '22728', '22726']

#リスト化した商品コードの商品名を調べる
for i in item_list:
    print(i, item_dict[i])
```

上までで相関がわかり、知見を得ることができたが  
ビジネスへの効果はまだ発揮できていない

以下からチューニング

```python
#アプリオリによる分析
#1回目より閾値を上げる
freq_items2 = apriori(basket_df, min_support=0.065, use_colnames=True)

#アソシエーションのルール抽出
a_rules2 = association_rules(freq_items2, metric='lift', min_threshold=1)

#リフト値でソート
a_rules2 = a_rules2.sort_values('lift', ascending=False).reset_index(drop=True)

#結果
display(a_rules2.head(10))
```

```python
#調査対象の集合
t_set = set([23254, 23256])

#1回目の分析freq_item1から該当行を抽出
idx1 = freq_items1['itemsets'].map(lambda x: not x.isdisjoint(t_set))
item1 = freq_items1[idx1]

#2回目の分析freq_item1から該当行を抽出
idx2 = freq_items2['itemsets'].map(lambda x: not x.isdisjoint(t_set))
item2 = freq_items2[idx2]

#結果確認
display(item1)
display(item2)
```

関係グラフの描画

```python
#対象とする関係を限定する
a = a_rules1.head(20)

#親ノードの抽出
#antecedentsは前提商品
ant = a['antecedents'].values
ant = [tuple(x) for x in ant]

#子ノードの抽出
#consequents帰結商品
con = a['consequents'].values
con = [tuple(x) for x in con]

#全ノードのリストアアップ
both = ant + con
both = list(set(both))

#結果の一部確認
print(both[:10])
```

関係グラフの構築

```python
# networkx のインポート
import networkx

# 関係グラフの初期化
s = networkx.DiGraph()
```

```python

# ノードの追加
for n in both:
    s.add_node(n)

# エッジの追加
for i in range(len(a)):
    item = a.loc[i]
    ant = tuple(item['antecedents'])
    con = tuple(item['consequents'])
    s.add_edge(ant, con)
```

```python
# グラフ描画
pos = networkx.spring_layout(s, k=0.6, seed=random_seed)

plt.figure(figsize=(8, 8))
networkx.draw_networkx_nodes(s, pos)
networkx.draw_networkx_edges(s, pos)
networkx.draw_networkx_labels(s, pos,
    horizontalalignment='left', 
    verticalalignment='center')
plt.axis('off')
plt.tight_layout()
plt.show()
```
